```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Task 1 : Party wise seats won

```python
seats_distribution_data = pd.read_csv("../data/seats-distribution.csv")
```

```python
seats_distribution_data.head()
```

```python
total_seats = seats_distribution_data["Won"].sum()
total_seats
```

```python
sns.set_theme(rc={'figure.figsize':(12,8.5)})
ax = sns.barplot(seats_distribution_data.loc[:,"Party Abbreviation":"Won"],x="Won", y="Party Abbreviation", hue="Party Abbreviation", estimator="sum", palette="dark")

for i in range(seats_distribution_data["Won"].shape[0]):
    ax.bar_label(ax.containers[i], fontsize=10)
```

```python
others = seats_distribution_data.loc[1:,"Won"].sum()
```

```python
pie_data = [240,others]
pie_keys = ['BJP','Others']

explode = [0.01, 0]
palette_color = sns.color_palette('pastel')
plt.figure().set_figheight(5)
plt.pie(pie_data, labels=pie_keys, colors=palette_color, autopct='%1.3f%%', explode=explode)
plt.show()
```

Insights:
- BJP had the most seats won with 240 seats out of 543
- BJP(being the leading party) vs all the others combined had a seat proportion ratio of 44.199% : 55.801%. This showcases that BJP does not have the clear cut majority in Lok Sabha
